```python
from bs4 import BeautifulSoup as bs
import requests
import csv
import pandas as pd
```

```python
url="https://merolagani.com/StockQuote.aspx"
def nepse(url,i,output):
  final_link=f"{url}/{i}"
  url = bs(final_link,'html.parser') 
  soup=requests.get(url)
  print(soup)
  code = soup.text
  lxml_code=bs(code, "lxml")
  table =lxml_code.table
  tag = table.find_all('tr')
  tag[0].text.split('\n')
  for i in range(len(tag)):
      datas = tag[i].text.split('\n')[1:-1]
      if datas[0]!='':
          output.append(datas)
  return output
output=[]
for j in range(0,2):
  nepse(url,j,output)
  output
  file = open('final.csv','w')
x=csv.writer(file)
for i in output:
    x.writerow(i)
df= pd.read_csv('/content/final.csv')
```
